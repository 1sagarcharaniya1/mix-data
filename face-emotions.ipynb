```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import glob
import os
from PIL import Image
import random
%matplotlib inline
```

```python
angry = glob.glob("/kaggle/input/face-expression-recognition-dataset/images/train/angry/*.jpg")
disgusted = glob.glob("/kaggle/input/face-expression-recognition-dataset/images/train/disgust/*.jpg")
fearful = glob.glob("/kaggle/input/face-expression-recognition-dataset/images/train/fear/*.jpg")
happy = glob.glob("/kaggle/input/face-expression-recognition-dataset/images/train/happy/*.jpg")
neutral = glob.glob("/kaggle/input/face-expression-recognition-dataset/images/train/neutral/*.jpg")
sad = glob.glob("/kaggle/input/face-expression-recognition-dataset/images/train/sad/*.jpg")
surprised = glob.glob("/kaggle/input/face-expression-recognition-dataset/images/train/surprise/*.jpg")
```

```python
angry_img = []

for i in angry:
    opn = Image.open(i)
    arry = np.array(opn)
    res = arry.reshape(2304)  
    angry_img.append(res)
    
new_ar = np.array(angry_img)
angry_df = pd.DataFrame(new_ar)
angry_df['Target'] = "angry"
```

```python
disgusted_img = []

for i in disgusted:
    opn = Image.open(i)
    arry = np.array(opn)
    res = arry.reshape(2304)  
    disgusted_img.append(res)
    
new_ar = np.array(disgusted_img)
disgusted_df = pd.DataFrame(new_ar)
disgusted_df['Target'] = "disgusted"
```

```python
fearful_img = []

for i in fearful:
    opn = Image.open(i)
    arry = np.array(opn)
    res = arry.reshape(2304)  
    fearful_img.append(res)
    
new_ar = np.array(fearful_img)
fearful_df = pd.DataFrame(new_ar)
fearful_df['Target'] = "fearful"
```

```python
happy_img = []

for i in happy:
    opn = Image.open(i)
    arry = np.array(opn)
    res = arry.reshape(2304)  
    happy_img.append(res)
    
new_ar = np.array(happy_img)
happy_df = pd.DataFrame(new_ar)
happy_df['Target'] = "happy"
```

```python
neutral_img = []

for i in neutral:
    opn = Image.open(i)
    arry = np.array(opn)
    res = arry.reshape(2304)  
    neutral_img.append(res)
    
new_ar = np.array(neutral_img)
neutral_df = pd.DataFrame(new_ar)
neutral_df['Target'] = "neutral"
```

```python
sad_img = []

for i in sad:
    opn = Image.open(i)
    arry = np.array(opn)
    res = arry.reshape(2304)  
    sad_img.append(res)
    
new_ar = np.array(sad_img)
sad_df = pd.DataFrame(new_ar)
sad_df['Target'] = "sad"
```

```python
surprised_img = []

for i in surprised:
    opn = Image.open(i)
    arry = np.array(opn)
    res = arry.reshape(2304)  
    surprised_img.append(res)
    
new_ar = np.array(surprised_img)
surprised_df = pd.DataFrame(new_ar)
surprised_df['Target'] = "surprised"
```

```python
df = pd.concat([angry_df,disgusted_df,fearful_df,happy_df,neutral_df,sad_df,surprised_df],ignore_index=True)
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df = df.sample(28821,replace=False).reset_index(drop=True)
```

```python
df
```

```python
x_train = df.iloc[:,:-1][:20000]
y_train = df.iloc[:,-1][:20000]

x_test = df.iloc[:,:-1][20000:]
y_test = df.iloc[:,-1][20000:]
```

```python
x_train_array = np.array(x_train) 
x_test_array = np.array(x_test)   
```

```python
x_train_array
```

```python
x_train_3d = x_train_array.reshape(20000,48,48)
x_test_3d = x_test_array.reshape(8821,48,48)
```

```python
i = 1680
image = x_train_3d[i]
label = y_train[i]
print(f'Label:- {label}')
plt.imshow(image);
```

```python
from sklearn.preprocessing import LabelBinarizer

enc = LabelBinarizer()

y_train = enc.fit_transform(y_train)
y_test = enc.fit_transform(y_test)

print(y_train.shape)
print(y_test.shape)
```

```python
y_train[1680]
```

```python
x_train_3d.shape[1:]
```

```python
enc.classes_
```

```python
from tensorflow.keras import models, layers, optimizers
```

```python
model = models.Sequential()

model.add(layers.Conv2D(32, (3, 3), activation="relu", input_shape=(48,48,1))),

model.add(layers.Conv2D(64, (3, 3), activation="relu"))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.1))

model.add(layers.Conv2D(128, (3, 3), activation="relu"))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.1))

model.add(layers.Conv2D(256, (3, 3), activation="relu"))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.1))


model.add(layers.Flatten())

model.add(layers.Dense(512, activation="relu"))
model.add(layers.Dropout(0.2))

model.add(layers.Dense(7, activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
x_train_3d_norm = x_train_3d.astype('float32')
x_test_3d_norm = x_test_3d.astype('float32')

x_train_3d_norm /= 255.0 
x_test_3d_norm /= 255.0
```

```python
x_train_3d_norm[0]
```

```python
x_train_final = x_train_3d_norm.reshape(x_train_3d_norm.shape[0], 48, 48)
x_test_final = x_test_3d_norm.reshape(x_test_3d_norm.shape[0], 48, 48)
```

```python
print(x_train_final.shape)
print(x_test_final.shape)
```

```python
y_train[0]
```

```python
model.fit(x_train_final, y_train, epochs=25)
```

```python
## 0.5784
## 0.9140
```

```python
model.evaluate(x_test_final, y_test)
```
